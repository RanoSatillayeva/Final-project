```python
df['date'] = pd.to_datetime(df['date'])

dau = df.groupby('date')['user_id'].nunique().reset_index()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
cohort = df[df['date'] == '2023-11-01']['user_id'].unique()
```

```python
day1_users = df[
    (df['date'] == '2023-11-02') &
    (df['user_id'].isin(cohort))
]['user_id'].nunique()
```

```python
retention = day1_users / len(cohort) * 100
print(retention)
```

```python
# общее количество пользователей
total_users = df['user_id'].nunique()

# пользователи, которые смотрели объявления
users_with_views = df[df['view_adverts'] > 0]['user_id'].nunique()

# конверсия
conversion = users_with_views / total_users * 100

print(f'Conversion: {conversion:.2f}%')
```

```python
avg_views = df['view_adverts'].sum() / df['user_id'].nunique()
print(avg_views)
```
